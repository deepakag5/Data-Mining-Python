```python
# Load the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the data(Data Source: Gapminder {http://www.gapminder.org})

gdp_percapita = pd.read_excel("GDPpercapitaconstant2000US.xlsx")
income_perperson = pd.read_excel("indicator gapminder gdp_per_capita_ppp.xlsx")
world_population = pd.read_excel("indicator gapminder population.xlsx")
co2emissions_percapita = pd.read_excel("indicator CDIAC carbon_dioxide_emissions_per_capita.xlsx")
co2emissions_cumulative = pd.read_excel("indicator CDIAC carbon_dioxide_cumulative_emissions.xlsx")
co2emissions_total = pd.read_excel("indicator CDIAC carbon_dioxide_total_emissions.xlsx")
```

```python
# Rename the first column in all dataframes

gdp_percapita.columns.values[0] = 'country'
income_perperson.columns.values[0] = 'country'
co2emissions_percapita.columns.values[0] = 'country'
co2emissions_cumulative.columns.values[0] = 'country'
co2emissions_total.columns.values[0] = 'country'
world_population.columns.values[0] = 'country'
```

```python
# Melt all the columns except country so that the year columns become individual rows

gdp_percapita_melt = pd.melt(gdp_percapita,id_vars=['country'], var_name='year',value_name='gdp_percapita')
income_perperson_melt = pd.melt(income_perperson,id_vars=['country'], var_name='year',value_name='income_perperson')
co2emissions_percapita_melt = pd.melt(co2emissions_percapita,id_vars=['country'], var_name='year',value_name='co2emissions_percapita')
co2emissions_cumulative_melt = pd.melt(co2emissions_cumulative,id_vars=['country'], var_name='year',value_name='co2emissions_cumulative')
co2emissions_total_melt = pd.melt(co2emissions_total,id_vars=['country'], var_name='year',value_name='co2emissions_total')
world_population_melt = pd.melt(world_population,id_vars=['country'], var_name='year',value_name='population')
```

```python
# Convert the year to numeric columns so that the dataframes can be merged

gdp_percapita_melt['year'] = pd.to_numeric(gdp_percapita_melt['year'])
income_perperson_melt['year'] = pd.to_numeric(income_perperson_melt['year'])
co2emissions_percapita_melt['year'] = pd.to_numeric(co2emissions_percapita_melt['year'])
co2emissions_cumulative_melt['year'] = pd.to_numeric(co2emissions_cumulative_melt['year'])
co2emissions_total_melt['year'] = pd.to_numeric(co2emissions_total_melt['year'])
world_population_melt['year'] = pd.to_numeric(world_population_melt['year'])
```

```python
# Check the dataframe
world_population_melt.dtypes
```

```python
# Perform inner join on the gdp_capita and income_perperson

gdp_income_population_data = gdp_percapita_melt.merge(income_perperson_melt,
                                                      on=['country','year']).merge(world_population_melt,on=['country','year'])
```

```python
co2emissions_data = co2emissions_percapita_melt.merge(co2emissions_cumulative_melt,
                                                      on=['country','year']).merge(co2emissions_total_melt,on=['country','year'])
```
